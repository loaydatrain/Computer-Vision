```
import cv2
import numpy as np 
import plotly.express as px
from google.colab.patches import cv2_imshow 
import matplotlib.pyplot as plt
```

#Loading and displaying

```
p1=cv2.imread("/content/drive/MyDrive/IIITH Assignments 3-2/CV/ASGN0/download (1).jpg")
p2=cv2.imread("/content/drive/MyDrive/IIITH Assignments 3-2/CV/ASGN0/download (2).jpg")
p3=cv2.imread("/content/drive/MyDrive/IIITH Assignments 3-2/CV/ASGN0/download (3).jpg")
p0=cv2.imread("/content/drive/MyDrive/IIITH Assignments 3-2/CV/ASGN0/download.jpg")

fig, sub=plt.subplots(2,2,figsize=(10,10))
sub[0][0].imshow(p1)
sub[0][0].axis('off')
sub[0][0].set_title("p1")

sub[0][1].imshow(p2)
sub[0][1].axis('off')
sub[0][1].set_title("p2")

sub[1][0].imshow(p3)
sub[1][0].axis('off')
sub[1][0].set_title("p3")

sub[1][1].imshow(p0)
sub[1][1].axis('off')
sub[1][1].set_title("p0")
```

#Fixed issue, convert to RGB

```
p0 = cv2.cvtColor(p0, cv2.COLOR_BGR2RGB)
p1 = cv2.cvtColor(p1, cv2.COLOR_BGR2RGB)
p2 = cv2.cvtColor(p2, cv2.COLOR_BGR2RGB)
p3 = cv2.cvtColor(p3, cv2.COLOR_BGR2RGB)

fig1, sub1=plt.subplots(2,2,figsize=(15,15))
sub1[0][0].imshow(p1)
sub1[0][0].axis('off')
sub1[0][0].set_title("p1")

sub1[0][1].imshow(p2)
sub1[0][1].axis('off')
sub1[0][1].set_title("p2")

sub1[1][0].imshow(p3)
sub1[1][0].axis('off')
sub1[1][0].set_title("p3")

sub1[1][1].imshow(p0)
sub1[1][1].axis('off')
sub1[1][1].set_title("p0")
```

#Convert to grayscale

```
r, g, b = p0[:,:,0], p0[:,:,1], p0[:,:,2]
gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

cv2_imshow(gray)
cv2.imwrite("asdf.jpg",gray)
```
